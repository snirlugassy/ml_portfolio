```python
import os
from glob import glob

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
S = glob("../experiments/*/test_share_ratio.txt")

S = []
for fp in glob("../experiments/*/test_share_ratio.txt"):
    with open(fp) as f:
        x = f.readlines()
        x = [float(_x) for _x in x]
        S.append(np.array(x))

S
```

```python
plt.figure(figsize=(20,5))
for s in S:
    plt.plot(s)

plt.ylim(-30,100)
plt.show()
```
